```python
import csv
from collections import Counter
from datetime import datetime

weekend_repo_counts = Counter()

with open('repositories.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)

    for row in reader:
        created_at = row.get('created_at', '')
        if created_at:
            created_date = datetime.fromisoformat(created_at[:-1])

            if created_date.weekday() in [5, 6]:
                user_login = row['login']
                weekend_repo_counts[user_login] += 1

top_users = weekend_repo_counts.most_common(5)

top_logins = [user[0] for user in top_users]

print(','.join(top_logins))
```

```python
import csv

def clean_company_name(company):
    if company:
        company = company.strip().upper()
        if company.startswith('@'):
            company = company[1:]
    return company

def clean_users_csv(input_file='users.csv', output_file='users.csv'):
    cleaned_users = []
    with open(input_file, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            row['company'] = clean_company_name(row['company'])
            row['hireable'] = 'true' if row['hireable'] == 'true' else 'false'
            row = {k: (v if v is not None else '') for k, v in row.items()}  # Replace None with empty string
            cleaned_users.append(row)

    with open(output_file, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=reader.fieldnames)
        writer.writeheader()
        writer.writerows(cleaned_users)

def clean_repos_csv(input_file='repositories.csv', output_file='repositories.csv'):
    cleaned_repos = []
    with open(input_file, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            row['has_projects'] = 'true' if row['has_projects'] == 'true' else 'false'
            row['has_wiki'] = 'true' if row['has_wiki'] == 'true' else 'false'
            row['license_name'] = row['license_name'] if row['license_name'] is not None else ''
            row = {k: (v if v is not None else '') for k, v in row.items()}  # Replace None with empty string
            cleaned_repos.append(row)

    with open(output_file, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=reader.fieldnames)
        writer.writeheader()
        writer.writerows(cleaned_repos)

if __name__ == "__main__":
    clean_users_csv()
    clean_repos_csv()
    print("CSV files cleaned and updated.")
```
